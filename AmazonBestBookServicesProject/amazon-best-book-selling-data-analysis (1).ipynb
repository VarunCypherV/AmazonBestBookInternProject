```python
#Book details based on its user ratings using groupby method
ratings=dataset[['Name','Price','User Rating','Reviews','Year','Author']]
ratings_groupby=ratings.groupby(['Name','Price','Reviews','Year','Author']).agg(np.max)
ratings_groupby.head(7) #visualize first 7 data
```

```python
#Book details based on its prices using groupby method
ratings1=dataset[['Name','Price','User Rating','Reviews','Year','Author']]
ratings_groupby1=ratings1.groupby(['Name','User Rating','Reviews','Year','Author']).agg(np.max)
ratings_groupby1.head(7) #visualize first 7 data
```

```python
#Count highest selling prices on books
count_ratings=ratings.groupby('Price').count()
count_ratings['perc_total']=round(count_ratings['User Rating']*100/count_ratings['User Rating'].sum(),1)
count_ratings['perc_total'].plot.bar()
```

The Highest selling books of price is 8 RS.

```python
#Count highest ratings on books
count_ratings1=ratings.groupby('User Rating').count()
count_ratings1['perc_total']=round(count_ratings1['Price']*100/count_ratings1['Price'].sum(),1)
count_ratings1['perc_total'].plot.bar()
```

Highest ratings got on book is 4.8
